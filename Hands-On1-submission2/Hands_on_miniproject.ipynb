```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
joke_setup = input("Enter a joke starter: ").strip()

output = generator(
    joke_setup,
    max_new_tokens=20,
    do_sample=True,
    temperature=0.7,
    truncation=True
)

print("\nJoke:")
print(joke_setup)
print(output[0]["generated_text"][len(joke_setup):].strip())
```
